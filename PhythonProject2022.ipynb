```python
#This is my second visualization as to what an early predictor of diabetes is. One of the biggest indicators that 
#a person may be diabetic with high glucose which inturn leads to higher insulin. 
#the pancreas produces insulin, a hormone that prompts cells to absorb blood sugar for energy or storage
#as you can see from the graph below the ones who have high glucose levels also have the higher insulin usually
#this is one thing patients must monitor for if they can potentially be diabetic

df.plot.scatter(x="Glucose", y="Insulin", alpha=0.5)
xlabel='Glucose'
plt.show()

#Now here are the results as you can see those with higher glucose and insulin are usually more 
#likely to be diabetic
sns.relplot(x='Glucose', y='Insulin', data=df, hue='Outcome')
plt.show()
```

```python
#This is my third visualization. In this we have all the categories that are significant factors in terms
#of if a patient has diabetes or not. I am showing this graph to emphasize the importance that glucose levels have
#on whether a patient is diabetic or not. As u can see all the other categories have scattered values however, when we 
#look at glucose level there is a common trend that higher the glucose the more likely a person's diabetic. 

cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction']

for col in cols :
        plt.figure(figsize=(4,4))         
        sns.lmplot(x="Age", y=col, hue = "Outcome", data=df)   
plt.show()
```

```python
#This is my fourth visualization. This was an interesting finding I found while conducting my research. In terms of age 
#being a factor in terms of if a patient is diabetic or not. What I found interesting is that patients around the 
#age of 45-55 are the ones that are most likely diabetic. This could be due to a variety of factors such as sample size,
#as the number of patients decreases as age does. Although this is the case, we see most of the diabetic patients
#come from this age group so as we age we need to be more cognizant.

plt.figure(figsize=(16,12))
cols = ['Age']
cnt=1
for col in cols :
        plt.subplot(3,3,cnt) 
        sns.histplot(x = col, data = df, hue = "Outcome")
        cnt= cnt+1
plt.tight_layout() 
plt.show()

#Group by aggregation to further prove the point and show how women from the age of 45-55 are most likely
#to get diabetes. This is shown as I only pull those 2 collums and make 'Age' the groupby and 'Outcome' the aggregate
df = pd.read_csv("diabetes.csv")
gk= df.groupby('Age').agg('Outcome')
gk.first()
```

```python
#This is my fifth visualization and I wanted to talk about the importance of BMI and diabetes. BMI is body mass index 
#Being overweight raises your risk for type 2 diabetes. If we look at the graph below we see a pretty obvious trend
#a patient with higher BMI is the most likely to be diabetic as you can see from the graph. This is to reiterate
#the importance of mentoring your weight as its a huge factor for being diabetic or not

fig = plt.figure(figsize=[15,4])
plt.subplot(1,2,2)
sns.histplot(data=df, hue='Outcome', x="BMI",kde=True)

plt.show()
```

```python
#This is my sixth visualization and here we are looking at how pregnancy in women impacts diabetes. Since this is 
#data in just women this is important to talk about. As we can see from the graph pregnancies do not have much
#of an impact on being diabetic. One interesting thing to look at is that once a women had up to 6 pregnancies the 
#risk for diabetes goes up. We can see this as the number of diabetic patients are greater than the non diabetic from
#the pregnancy range of 6-9. This is something not many women take into consideration as most women are at much greater
#risk of being diabetic after 6+ pregnancies

plt.figure(figsize=(16,12))
cols = ['Pregnancies']
cnt=1
for col in cols :
        plt.subplot(3,3,cnt) 
        sns.histplot(x = col, data = df, hue = "Outcome")
        cnt= cnt+1
plt.tight_layout() 
plt.show()

#Groupby aggregation to further visualize the data and show how 6+ pregnancies put u at much higher risk of diabetes
#done by isolating pregnancy number and outcome
df = pd.read_csv("diabetes.csv")
gk= df.groupby('Pregnancies').agg('Outcome')
gk.first()
```
